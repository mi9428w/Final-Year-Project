```python
orders = pd.read_csv("olist_orders_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")

# Only delivered orders
orders = orders[orders["order_status"] == "delivered"]

# Merge to get customer_unique_id
orders = orders.merge(customers[["customer_id", "customer_unique_id"]],
                      on="customer_id",
                      how="left")

print(orders.shape)
orders.head()
```

```python
order_items = pd.read_csv("olist_order_items_dataset.csv")

orders = orders.merge(
    order_items[["order_id", "price", "freight_value"]],
    on="order_id",
    how="left"
)

print(orders.shape)
orders.head()
```

```python
customer_data = orders.groupby("customer_unique_id").agg(
    total_orders=("order_id", "nunique"),
    total_spent=("price", "sum"),
    avg_freight=("freight_value", "mean"),
    first_purchase=("order_purchase_timestamp", "min"),
    last_purchase=("order_purchase_timestamp", "max")
).reset_index()

print(customer_data.shape)
customer_data.head()
```

```python
# make sure dates are datetime
customer_data["first_purchase"] = pd.to_datetime(customer_data["first_purchase"])
customer_data["last_purchase"]  = pd.to_datetime(customer_data["last_purchase"])

# recency (days since last purchase, relative to latest date in dataset)
customer_data["recency_days"] = (
    customer_data["last_purchase"].max() - customer_data["last_purchase"]
).dt.days

customer_data[["customer_unique_id", "last_purchase", "recency_days"]].head()
customer_data["recency_days"] = (
    customer_data["last_purchase"].max() - customer_data["last_purchase"]
).dt.days

customer_data[["customer_unique_id", "recency_days"]].head()
```

```python
threshold = customer_data["total_spent"].median()

customer_data["treatment"] = (
    customer_data["total_spent"] > threshold
).astype(int)

customer_data["treatment"].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

# covariates
X = customer_data[["total_orders","avg_freight","recency_days"]]

# treatment
y = customer_data["treatment"]

model = LogisticRegression(max_iter=1000)
model.fit(X,y)

customer_data["propensity_score"] = model.predict_proba(X)[:,1]

customer_data[["customer_unique_id","propensity_score"]].head()
```

```python
customer_data["propensity_score"].describe()
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

#divided to treated and control group
treated = customer_data[customer_data["treatment"] ==1]
control = customer_data[customer_data["treatment"]==0]

#propensity score array
treated_scores = treated["propensity_score"].values.reshape(-1,1)
control_scores = control["propensity_score"].values.reshape(-1,1)

#nearest neighbor matching
nn = NearestNeighbors(n_neighbors=1)
nn.fit(control_scores)

distance, indices = nn.kneighbors(treated_scores)

#matched control customers
matched_control = control.iloc[indices.flatten()]

#matched dataset
matched_data = pd.concat([treated.reset_index(drop=True),
                          matched_control.reset_index(drop=True)])

matched_data.head()
```

```python
matched_data.groupby("treatment")[["total_orders","avg_freight","recency_days"]].mean()
```

```python
matched_data.groupby("treatment")["total_spent"].mean()
```

```python
import matplotlib.pyplot as plt

treated_scores = matched_data[matched_data["treatment"]==1]["propensity_score"]
control_scores = matched_data[matched_data["treatment"]==0]["propensity_score"]

plt.hist(treated_scores, bins=30, alpha=0.5, label="Treated")
plt.hist(control_scores, bins=30, alpha=0.5, label="Control")

plt.xlabel("Propensity Score")
plt.ylabel("Frequency")
plt.title("Propensity Score Distribution After Matching")
plt.legend()

plt.show()
```

```python
matched_data.boxplot(column="total_spent", by="treatment")

plt.title("Spending Comparison Between Groups")
plt.suptitle("")
plt.xlabel("Treatment Group")
plt.ylabel("Total Spending")

plt.show()
```

```python
#Creating customer segments based on propensity score. Customers are divided into three groups Low, Medium and High Propensity
customer_data["propensity_segment"] = pd.cut(
    customer_data["propensity_score"],
    bins=[0,0.4,0.7,1],
    labels=["Low","Medium","High"]
)
customer_data.head()
```

```python
#counting how many customers are in each propensity segment
customer_data["propensity_segment"].value_counts()
```

```python
#Calculating the average spending for each segment
customer_data.groupby("propensity_segment")["total_spent"].mean()
```

```python
# This step simulates how an e-commerce system could recommend
# additional products based on the customer's propensity segment
# and the items currently in their shopping basket.
```

```python
# Creating recommendation rules based on basket items

basket_rules = {
    "Shoes": ["Socks", "Shoe Cleaner", "Bag"],
    "Watch": ["Bracelet", "Premium Wallet", "Sunglasses"],
    "Bag": ["Wallet", "Sunglasses"],
    "T-shirt": ["Cap", "Jeans"],
    "Headphones": ["Bluetooth Speaker", "Laptop Bag"],
    "Sunglasses": ["Cap", "Watch"],
    "Laptop Bag": ["Mouse", "Headphones"],
    "Wallet": ["Belt", "Watch"]
}

basket_rules
```

```python
#Getting candidate recommendations based on basket items

candidate_products =[]

for item in basket:


    if item in basket_rules:
        candidate_products.extend(basket_rules[item])

candidate_products = list(set(candidate_products))

candidate_products
```

```python
# Final recommendation using customer segment and basket suggestions

import random

if segment == "High":
    recommendation = random.choice(candidate_products)

elif segment == "Medium":
    recommendation = random.choice(candidate_products)

else:
    recommendation = "Discount " + random.choice(candidate_products)

print("Customer Segment:", segment)
print("Basket Items:", basket)
print("Candidate Products:", candidate_products)
print("Final Recommended Product:", recommendation)
```
