```python
orders = pd.read_csv("olist_orders_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")

# Only delivered orders
orders = orders[orders["order_status"] == "delivered"]

# Merge to get customer_unique_id
orders = orders.merge(customers[["customer_id", "customer_unique_id"]],
                      on="customer_id",
                      how="left")

print(orders.shape)
orders.head()
```

```python
order_items = pd.read_csv("olist_order_items_dataset.csv")

orders = orders.merge(
    order_items[["order_id", "price", "freight_value"]],
    on="order_id",
    how="left"
)

print(orders.shape)
orders.head()
```

```python
customer_data = orders.groupby("customer_unique_id").agg(
    total_orders=("order_id", "nunique"),
    total_spent=("price", "sum"),
    avg_freight=("freight_value", "mean"),
    first_purchase=("order_purchase_timestamp", "min"),
    last_purchase=("order_purchase_timestamp", "max")
).reset_index()

print(customer_data.shape)
customer_data.head()
```

```python
# make sure dates are datetime
customer_data["first_purchase"] = pd.to_datetime(customer_data["first_purchase"])
customer_data["last_purchase"]  = pd.to_datetime(customer_data["last_purchase"])

# recency (days since last purchase, relative to latest date in dataset)
customer_data["recency_days"] = (
    customer_data["last_purchase"].max() - customer_data["last_purchase"]
).dt.days

customer_data[["customer_unique_id", "last_purchase", "recency_days"]].head()
customer_data["recency_days"] = (
    customer_data["last_purchase"].max() - customer_data["last_purchase"]
).dt.days

customer_data[["customer_unique_id", "recency_days"]].head()
```

```python
threshold = customer_data["total_spent"].median()

customer_data["treatment"] = (
    customer_data["total_spent"] > threshold
).astype(int)

customer_data["treatment"].value_counts()
```
